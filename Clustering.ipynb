```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el nuevo dataset en Python
df = pd.read_csv("C:/Users/F16216/Documents/Milagros/Maestría/DMEYF/competencia_01_clusters.csv")

```

## Algunas transformaciones para la visualización

```python
df['total_mpayroll']=df['mpayroll']+df['mpayroll2'] #monto total de remuneración
df['mdeudas']=df['mprestamos_personales']+df['mprestamos_hipotecarios']+df['mprestamos_prendarios']
df['cdeudas']=df['cprestamos_personales']+df['cprestamos_hipotecarios']+df['cprestamos_prendarios']
variables_transacciones=['ctarjeta_debito_transacciones','ctarjeta_visa_transacciones','ctarjeta_master_transacciones','cpayroll_trx','cpayroll2_trx','cforex',
                         'ccallcenter_transacciones','chomebanking_transacciones','ccajas_transacciones','ccajas_otras','catm_trx',
                         'catm_trx_other','cmobile_app_trx']
#Creo mi variable de cantidad de transacciones
df['cantidad_transacciones'] = df[variables_transacciones].sum(axis=1)
#Me gustaría sacar mi top 1 para ver si el cliente ha ido variando el comportamiento a lo largo de meses.
df['top1_transacciones'] = df[variables_transacciones].idxmax(axis=1)
df.loc[df['cantidad_transacciones'] == 0, 'top1_transacciones'] = 'ninguna'
df['tarjeta_saldototal']=df['Visa_msaldototal']+df['Master_msaldototal']
```

```python
df['cluster']=df['cluster'].map({'cluster_1':'Zeus',
                                 'cluster_2':'Atena',
                                 'cluster_3':'Poseidon',
                                 'cluster_4':'Hefesto',
                                 'cluster_5':'Hades'})
```

```python
df_filtered = df[df['foto_mes'] <= 202105]
```

## Visualizaciones

```python
import warnings

# Suprimir advertencias
warnings.filterwarnings("ignore")
```

```python
# Mapeo de clusters a colores específicos
colores_cluster = {
    'Zeus': "#A3C4F3",    # Azul claro
    'Hefesto': "#FF9F80", # Naranja suave
    'Atena': "#FFE6A7",   # Amarillo pálido
    'Poseidon': "#88D9D1",# Verde agua
    'Hades': "#B56576"    # Rojo oscuro
}
```

```python
# Contar la cantidad de registros por categoría en 'cluster'
clusters_counts = df['cluster'].value_counts()

clusters = clusters_counts.values 
nombres = df['cluster'].value_counts().index
colores = [colores_cluster[nombre] for nombre in nombres]

# Función para mostrar % y cantidad
def autopct_format(values):
    def inner_autopct(pct):
        total = sum(values)
        absolute = int(pct/100.*total)
        return f'{pct:.1f}%\n({absolute})'
    return inner_autopct

# Crear el gráfico de pastel sin etiquetas
fig, ax = plt.subplots()
ax.pie(clusters, autopct=autopct_format(clusters), 
       colors=colores, startangle=90, textprops={'fontsize': 12})
ax.axis('equal')  # Asegurar que el gráfico sea circular

# Agregar leyenda al costado
ax.legend(nombres, title="Segmentos", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), fontsize=12)

# Mostrar gráfico
plt.show()
```

```python
from scipy import stats
import numpy as np

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x='mes', y='ctrx_quarter', hue='cluster', data=df_filtered, palette=colores_cluster, errorbar=None)

# Añadir etiquetas y título
plt.xlabel('Periodo')
plt.ylabel('Movimientos voluntarios (Q)')
plt.title('Cantidad de movimientos voluntarios por Periodo y Segmento')

# Calcular el promedio de ctrx_quarter por cada mes
mean_values = df_filtered.groupby('mes')['ctrx_quarter'].mean().reset_index()

# Convertir 'mes' a valores numéricos si no lo está
mean_values['mes_numeric'] = pd.to_numeric(mean_values['mes'], errors='coerce')

# Ajustar una regresión lineal usando los valores medios
slope, intercept, r_value, p_value, std_err = stats.linregress(mean_values['mes_numeric'], mean_values['ctrx_quarter'])

# Crear la línea de tendencia
x_vals = np.array(mean_values['mes_numeric'])


# Calcular los valores y para la línea de tendencia
y_vals = intercept + slope * x_vals

# Graficar la línea de tendencia usando los valores de mes directamente
plt.plot(mean_values['mes']-1, y_vals, color='violet', label='Tendencia', linewidth=2)

# Mostrar la leyenda, grid y el gráfico
plt.grid(True)
plt.legend(title='Segmentos')
plt.show()
```

```python
# Calcular la suma o conteo de transacciones por cluster
transacciones_por_cluster = df_filtered.groupby('cluster')['cantidad_transacciones'].mean().reset_index()

# Crear el gráfico de barras usando seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='cluster', y='cantidad_transacciones', data=transacciones_por_cluster, palette=colores_cluster)  # Usar una paleta cualitativa similar a la imagen

# Etiquetas y título
plt.xlabel('Segmento')
plt.ylabel('Cantidad de Transacciones')
plt.title('Cantidad de Transacciones por Segmentos')

# Mostrar el gráfico
plt.show()
```

```python
# Agrupar por cluster y top1_transacciones, y contar cuántas veces se repite cada transacción
top_transacciones = df_filtered.groupby(['cluster', 'top1_transacciones']).size().reset_index(name='count')

# Obtener las 5 transacciones más comunes para cada cluster
top3_transacciones_por_cluster = top_transacciones.groupby('cluster').apply(lambda x: x.nlargest(3, 'count')).reset_index(drop=True)

# Visualizar en un gráfico de barras apiladas o separado por clusters
plt.figure(figsize=(10, 6))
sns.barplot(x='cluster', y='count', hue='top1_transacciones', data=top3_transacciones_por_cluster, palette='tab10')
plt.title('Top 3 Transacciones por Cluster')
plt.xlabel('Segmentos')
plt.ylabel('Cantidad de Transacciones')
plt.legend(title='Transacción')
plt.show()
```

```python
# Calcular la suma o conteo de transacciones por cluster
deudas_por_cluster = df_filtered.groupby('cluster')['mdeudas'].sum().reset_index()

# Crear el gráfico de barras usando seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='cluster', y='mdeudas', data=deudas_por_cluster, palette=colores_cluster)  # Usar una paleta cualitativa similar a la imagen

# Etiquetas y título
plt.xlabel('Segmento')
plt.ylabel('Monto adeudado en concepto de préstamos')
plt.title('Deudas por Segmento')

# Mostrar el gráfico
plt.show()
```
