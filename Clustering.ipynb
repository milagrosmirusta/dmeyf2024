```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el nuevo dataset en Python
df = pd.read_csv("C:/Users/F16216/Documents/Milagros/Maestría/DMEYF/competencia_01_clusters.csv")

```

```python
df=df.loc[df['cluster'].notna()]
df.head()
```

```python
df['mes'] = df['foto_mes'] % 10  # Esto toma el último dígito del número
```

```python
df.mes.value_counts()
```

```python
df.describe()
```

## Algunas transformaciones para la visualización

```python
df['total_mpayroll']=df['mpayroll']+df['mpayroll2'] #monto total de remuneración
df['mdeudas']=df['mprestamos_personales']+df['mprestamos_hipotecarios']+df['mprestamos_prendarios']
df['cdeudas']=df['cprestamos_personales']+df['cprestamos_hipotecarios']+df['cprestamos_prendarios']
variables_transacciones=['ctarjeta_debito_transacciones','ctarjeta_visa_transacciones','ctarjeta_master_transacciones','cpayroll_trx','cpayroll2_trx','cforex',
                         'ccallcenter_transacciones','chomebanking_transacciones','ccajas_transacciones','ccajas_otras','catm_trx',
                         'catm_trx_other','cmobile_app_trx']
#Creo mi variable de cantidad de transacciones
df['cantidad_transacciones'] = df[variables_transacciones].sum(axis=1)
#Me gustaría sacar mi top 1 para ver si el cliente ha ido variando el comportamiento a lo largo de meses.
df['top1_transacciones'] = df[variables_transacciones].idxmax(axis=1)
df.loc[df['cantidad_transacciones'] == 0, 'top1_transacciones'] = 'ninguna'
df['tarjeta_saldototal']=df['Visa_msaldototal']+df['Master_msaldototal']
```

```python
df_filtered = df[df['foto_mes'] <= 202105]
```

```python
# Contar la cantidad de registros por categoría en 'cluster'
clusters_counts = df['cluster'].value_counts()


clusters = clusters_counts.values 
nombres = ["1","2","3","4", "5"]
colores = ["#1A75BB", "#00C4CC", "#FF1654", "#28D8A1", "#FFE600"]
plt.pie(clusters, labels=nombres, autopct="%0.1f %%", colors=colores, 
      startangle=90, textprops={'fontsize': 12})
plt.axis("equal")  # Asegurar que el gráfico sea circular
plt.show()
```

```python
# Calcular la suma o conteo de transacciones por cluster
transacciones_por_cluster = df_filtered.groupby('cluster')['cantidad_transacciones'].sum().reset_index()

# Crear el gráfico de barras usando seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='cluster', y='cantidad_transacciones', data=transacciones_por_cluster, palette=colores)  # Usar una paleta cualitativa similar a la imagen

# Etiquetas y título
plt.xlabel('Cluster')
plt.ylabel('Cantidad de Transacciones')
plt.title('Cantidad de Transacciones por Cluster')

# Mostrar el gráfico
plt.show()
```

```python
# Agrupar por cluster y top1_transacciones, y contar cuántas veces se repite cada transacción
top_transacciones = df_filtered.groupby(['cluster', 'top1_transacciones']).size().reset_index(name='count')

# Obtener las 5 transacciones más comunes para cada cluster
top5_transacciones_por_cluster = top_transacciones.groupby('cluster').apply(lambda x: x.nlargest(5, 'count')).reset_index(drop=True)

# Visualizar en un gráfico de barras apiladas o separado por clusters
plt.figure(figsize=(10, 6))
sns.barplot(x='cluster', y='count', hue='top1_transacciones', data=top5_transacciones_por_cluster, palette='tab10')
plt.title('Top 5 Transacciones por Cluster')
plt.xlabel('Cluster')
plt.ylabel('Cantidad de Transacciones')
plt.legend(title='Transacción')
plt.show()
```

```python
# Usar FacetGrid para crear un gráfico separado para cada cluster
g = sns.FacetGrid(df_filtered, col="cluster", height=4, aspect=1)
g.map(sns.scatterplot, "total_mpayroll", "tarjeta_saldototal")
g.map(sns.regplot, "total_mpayroll", "tarjeta_saldototal", scatter=False, color="red")

# Etiquetas y título
g.set_axis_labels("Total Payroll (Dinero depositado)", "Saldo Total en Tarjeta")
g.fig.suptitle('Relación entre Saldo de Tarjeta y Dinero Depositado por Cluster', y=1.05)
plt.show()
```

```python
# Calcular la suma o conteo de transacciones por cluster
deudas_por_cluster = df_filtered.groupby('cluster')['mdeudas'].sum().reset_index()

# Crear el gráfico de barras usando seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='cluster', y='mdeudas', data=deudas_por_cluster, palette=colores)  # Usar una paleta cualitativa similar a la imagen

# Etiquetas y título
plt.xlabel('Cluster')
plt.ylabel('Monto adeudado')
plt.title('Deudas por Cluster')

# Mostrar el gráfico
plt.show()
```
